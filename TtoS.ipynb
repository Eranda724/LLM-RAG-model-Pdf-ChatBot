```python
import gradio as gr
from gtts import gTTS
import os
import uuid
```

```python
def text_to_speech(text):
    if not text.strip():
        return "Please enter some text" 
    
    # Generate a unique filename
    filename = f"temp_{uuid.uuid4()}.mp3"
    
    # Convert text to speech
    tts = gTTS(text=text, lang='en')
    tts.save(filename)
    
    return filename, "Audio generated successfully!"

interface = gr.Interface(
    fn=text_to_speech,
    inputs=gr.Textbox(lines=4, label="Enter Text"),
    outputs=[gr.Audio(label="Generated Audio", type="filepath")
             , gr.Textbox(label="Status")],
    title="Text to Speech Converter",
    description="Enter text to convert it into speech. The audio will be generated and can be downloaded.",
)   

interface.launch()
```
