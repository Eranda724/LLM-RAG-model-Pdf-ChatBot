```python
import gradio as gr
from transformers import pipeline

pipe = pipeline("automatic-speech-recognition", model="openai/whisper-small")
```

```python
home_state = {
    "temperature": 18,
    "lights": "off",
    "ac": "off",
    "heater": "off"
}
```

```python
def process_command(command):
    command = command.lower()
    action_result = " "
    if "turn on the light" in command or "lights on" in command:
        home_state["lights"] = "on"
        action_result = "Lights turned on"
    elif "turn off the light" in command or "lights off" in command:
        home_state["lights"] = "off"
        action_result = "Lights turned off"
    elif "increase temperature" in command:
        home_state["temperature"] += 1
        action_result = f"Temperature increased to {home_state['temperature']} C"
    elif "decrease temperature" in command:
        home_state["temperature"] -= 1
        action_result = f"Temperature decreased to {home_state['temperature']} C"
    elif "turn on air conditioner" in command or "air conditioner on" in command:
        home_state["ac"] = "on"
        action_result = "AC turned on"
    elif "turn off air conditioner" in command or "air conditioner off" in command:
        home_state["ac"] = "off"
        action_result = "AC turned off"
    elif "turn on heater" in command or "heater on" in command:
        home_state["heater"] = "on"
        action_result = "Heater turned on"
    elif "turn off heater" in command or "heater off" in command:
        home_state["heater"] = "off"
        action_result = "Heater turned off"
    else:
        action_result = "Command not recognized"
    return (
        action_result,
        f"{home_state['temperature']} C",
        home_state["lights"],
        home_state["ac"],
        home_state["heater"]
    )
```

```python
def handle_audio(audio):
    if isinstance(audio, dict) and "name" in audio:
        audio = audio["name"]
    result = pipe(audio)
    command = result["text"]
    return process_command(command)
```

```python
with gr.Blocks(theme=gr.themes.Soft()) as demo:
    gr.Markdown("Command Voice Recognition System for Smart Home Automation")
    with gr.Row():
        audio_input = gr.Audio(type="filepath", label="Speak a Command")
        action_result = gr.Textbox(label="Action Result")
    run_button = gr.Button("Execute Voice Command")
    gr.Markdown("Smart Home Automation Dashboard")
    with gr.Row():
        temp_display = gr.Textbox(label="Temperature", value=f"{home_state['temperature']} C")
        lights_display = gr.Textbox(label="Lights", value=home_state["lights"])
        ac_display = gr.Textbox(label="AC", value= home_state["ac"])
        heater_display = gr.Textbox(label="Heater", value=home_state["heater"])
    run_button.click(
        fn=handle_audio,
        inputs=[audio_input],
        outputs=[action_result, temp_display, lights_display, ac_display, heater_display]
    )
demo.launch()
```
